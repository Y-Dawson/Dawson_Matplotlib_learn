```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from cycler import cycler
```

使用Style sheets

style package添加了对容易切换绘图的style的支持,这些样式和matplotlib rc文件具有相同的参数
mplt提供了很多预定义的样式

```python
#这里添加了ggplot的样式
plt.style.use('ggplot')
data = np.random.randn(50)
```

```python
print(plt.style.available)#列出预定义的样式
```

可以自定义样式并且通过style.use打开path or url来使用style sheet.
如果您将<style-name>.mplstyle文件添加 到mpl_configdir/stylelib，则可以通过调用来重复使用自定义样式表style.use(<style-name>)。默认情况下 mpl_configdir应该是~/.config/matplotlib，但是您可以检查您的位置matplotlib.get_configdir(); 您可能需要创建此目录。您还可以通过设置MPLCONFIGDIR环境变量来更改matplotlib在其中查找stylelib /文件夹的目录，请参见matplotlib配置和缓存目录位置。

请注意，mpl_configdir/stylelib如果样式具有相同的名称，则自定义样式表将覆盖由matplotlib定义的样式表。

组合Style

Style sheets被设计为可以被整合的,因此，您可以拥有一个可自定义颜色的样式表，以及一个可以更改演示文稿元素大小的单独样式表。通过传递样式列表，可以轻松地组合这些样式
import matplotlib.pyplot as plt
plt.style.use(['dark_background', 'presentation'])

#注意，右边的样式将覆盖左边的样式已经定义的值。

临时造型

如果您只想为特定的代码块使用样式，而又不想更改全局样式，则样式包提供了一个上下文管理器，用于将更改限制在特定范围内。为了隔离样式更改，您可以编写如下内容：

```python
with plt.style.context('dark_background'):
    plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

动态rc设置

您还可以在python脚本中动态更改默认的rc设置，或者从python shell交互式更改默认的rc设置。所有rc设置都存储在名为的类似于字典的变量中matplotlib.rcParams，该变量对于matplotlib软件包是全局的。rcParams可以直接修改，例如

```python
mpl.rcParams['lines.linewidth'] = 2
mpl.rcParams['lines.linestyle'] = '--'
plt.plot(data)
```

```python
#请注意，为了更改总体的plot颜色，您必须更改轴的 prop_cycle属性

mpl.rcParams['axes.prop_cycle'] = cycler(color=['black', 'b', 'b', 'y'])
plt.plot(data)  # first color is red
```

```python
#Matplotlib还提供了一些方便的功能来修改rc设置。该matplotlib.rc()命令可用于通过关键字参数一次修改单个组中的多个设置：

mpl.rc('lines', linewidth=4, linestyle='-.')
plt.plot(data)
```
